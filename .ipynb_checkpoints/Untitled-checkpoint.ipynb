```python
from sklearn import tree
import pandas as pd
import numpy as np
import os
```

```python
# Read the csv file into a pandas DataFrame
df = pd.read_csv('Resources/clean_genre.csv')
df.head()
```

```python
df = df.rename(columns = {"Total Nominations":"Oscar_Nominations"})

df['Oscar_Nominations'].loc[df['Oscar_Nominations'] >= 1] = 1

df.head()
```

```python
clean = df[['Oscar_Nominations','movie score', 'budget', 'gross','runtime', 'votes','star avg score', 'director avg score','writer avg score','genre avg score', 'company avg score', 'country avg score']]
clean.head(3)
```

```python
data = clean.drop('Oscar_Nominations', axis=1)
feature_names = data.columns
data.head()
```

```python
target = clean['Oscar_Nominations']
target_names = ["No", "Yes"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
sorted(zip(rf.feature_importances_, feature_names), reverse=True)
```

```python
# Predict the nominations(yes or no) of the new data point
predictions = rf.predict(X_test)
print("Classes are either 0 (No) or 1 (Yes)")
```

```python
predictions = rf.predict(X_test)
pd.DataFrame({"Prediction": predictions, "Actual": y_test})
```
